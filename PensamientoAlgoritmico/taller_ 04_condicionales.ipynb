> __1. Introducir el bloque__
- Existe un bloque en forma de paralelepípedo (todas sus seis caras son rectangulares), el cual mide $a$ unidades de ancho, $b$ de largo y $c$ de profundidad. También, hay un orificio circular de radio $r$. Diseñe un algoritmo que determine de cuántas formas distintas se puede introducir de manera perpendicular el bloque en el orificio. Es decir al momento de probar con cada cara del bloque con cuántas si se puede?


```python
# primer punto.
a=float(input(print("digíte el valor del ancho del paralelepipedo ")))
b=float(input(print("digíte el valor del largo del paralelepipedo ")))
c=float(input(print("digíte el valor de la profundidad del paralelepipedo ")))
r=float(input(print("digíte el valor del radio del orificio ")))

d1=((a**2)+(b**2))**0.5
d2=((a**2)+(c**2))**0.5
d3=((c**2)+(b**2))**0.5

d=r*2


if d1<d:
    print("cabe con la cara de los lados A & B.")
else:
    print("no cabe con la cara de A & B.")
if d2<d:
    print("cabe con la cara de los lados A & C.")
else:
    print("no cabe con la cara de A & C.")
if d3<d:
    print("cabe con la cara de los lados B & C.")
else:
    print("no cabe con la cara de B & C.")

if d1<d and d2<d and d3<d:
    print("el paralelepipedo cabe por todas las caras. :3")
else: 
    print("el paralelepipedo solo cabe por algunas caras <{:{v ")
```
