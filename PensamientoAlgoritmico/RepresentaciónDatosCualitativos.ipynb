## Simular una encuesta de operadores móviles preferidos por 50 personas en Colombia


```r
x = sample(c("Claro", "Tigo", "Movistar", "Avantel", "Virgin Mobile", "Wom"),size=50,replace=TRUE)
x
```

## Tabla de frecuencias 

```r
table(x)
# crea una tabla :u
```

```r
#tabla de frecuencias relativas
table(x)/length(x)
```

## Diagrama de Barras 

```r
# crea un diagrama de barras :u
barplot(table(x), main="Preferencia de Operadores en Colombia", col=c("blue", "red", "green", "yellow", "purple", "orange"))
```

## Diagrama de torta

```r
pie(table(x), main="Preferencia de Operadores en Colombia", col=c("#FF4500", "#B22222", "dodgerblue3", "deepskyblue1", "purple", "#68228B"))
```
