4. Hallar el promedio de una lista de valores. El algoritmo debe ignorar los valores nulos (`none`) si existen

```python
#4to punto
print('CUARTO PUNTO')
def promedio(lista):
    acum = 0
    cont=0
    for i in lista:
        if i!=None:
            acum = acum+i
            cont=cont+1
    print('la suma de los valores es:',acum,', la cantidad de valores es:',cont,'y el promedio de todos los valores es:',acum/cont)

L4=[3.5,4,3.5,5,2,2.5,3]
promedio(L4)
```

5. Dada una lista de valores, halle la desviación estándar. El algoritmo debe ignorar los valores nulos (`none`) si existen

```python
#5to punto
print('QUINTO PUNTO')
L5=[5,15,12,18,28]
def desviacion(w):
    acum= 0
    cont=0
    valor=0
    for i in w:
        if i!=None:
            acum=acum+i
            cont=cont+1
    promedio=acum/cont
    print('el promedio es:',promedio)
    for k in range(0,len(w)):
        valor=((w[k]-promedio)**2)+valor
    desviacion=(valor/(cont-1))**0.5
    print('la desviación estándar es igua a:',desviacion)
    
desviacion(L5)
```
