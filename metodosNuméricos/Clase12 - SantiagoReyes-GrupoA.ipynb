```python
import sympy as sp
from sympy import pi, I, E, oo
```

```python
## Evaluación numérica
#Ingresar números.

sp.N(1 + pi)    # El resultado es un numero decimal de 14 digitos 
sp.N(pi, 50)    #Produce el numero pi con un redondeo de 50 caracteres
# para mezclar simbolos y números
# Definimos un simbolo
x = sp.Symbol('x')
# considere una operación 
(1+x).evalf(10) #Realiza la suma de los dos simbolos con una precision de 10 digitos
```

```python
# Evaluación de funciones tipo lambda
expr = sp.sin(pi*x*sp.exp(x))       #función f(x) = sin (pi*x*exp(x))
expr_func = sp.lambdify(x,expr)
expr_2 = expr_func(1.0)
```

```python
# Calculo -> derivadas.

# defininción de funciones

f=sp.Function('F')(x)       #f(x)
df = sp.diff(f,x)           #Primera derivada de f
d2f = sp.diff(f,x,2)        #Segunda derivada de f

#Derivadas mixtas
y = sp.Symbol('y')
g = sp.Function('g')(x,y)   #Funcion multivariable
dfxy = g.diff(x,y)

dfxy

# ejemplo

expr = x**4+ x**3 + x**2 + x + 1
expr.diff(x)                #Obtención de la primera derivada.
expr.diff(x,2)              #Obtención de la segunda derivada.

# Ejemplo.
expr = sp.sin(x * y)*sp.cos(x/2)        #f(x) = sin(xy)*cos(0.5x)
expr.diff(x)                            # Derivada respecto a x
expr.diff(y)                            # Derivada respecto a y

# Ejemplo.
# represente la derivada solo simbolica
d = sp.Derivative(sp.exp(sp.cos(x)),x)
# Para evaluar la derivada simbolica usamos
d.doit()
```

```python
# Calculo -> integrales.
# Definimos la funcion.
f = sp.Function('f')(x)
a,b,x,y = sp.symbols('a,b,x,y')
# para integrar F usamos integrate
sp.integrate(f)     # integral indefinida
sp.integrate(f,(x,a,b))     # integral indefinida
# Ejemplo.
sp.integrate(sp.sin(x))     # Integral indefinida
sp.integrate(sp.cos(x))     # Integral indefinida

#Integrales definidas
sp.integrate(sp.sin(x),(x,a,b))
sp.integrate(sp.cos(x),(x,a,b))

#Ejemplo
sp.integrate(sp.exp(-x**2),(x,0,oo))

#Ejemplo.
a,b,c = sp.symbols('a,b,c', positive = True)
sp.integrate(a*sp.exp(-((x-b)/c)**2),(x,-oo,oo))

#Ejemplo integrales explicitas
expr = sp.sin(x*sp.exp(y))          # f(x)= sin(xexp(y))
sp.integrate(expr,x)                # derivada con respecto a x
sp.integrate(expr,y)                # derivada con respecto a y
sp.integrate(expr,y,y)              # derivada con respecto a x,y

#Ejemplo
expr = (x+y)**2
sp.integrate(expr,x,y)              #Derivada con respecto a x, y indefinida.
sp.integrate(expr,(x,0,1),(y,0,1))  #Derivada con respecto a x, y definida.
```

```python
# Calculo -> series.
# Definición de funcion.
f = sp.Function('f')(x)
sp.series(f,x)                  #Serie de McClaurin de f(x)
# Cambiar el punto
x0=sp.Symbol('x0')
f.series(x,0,8)                 #serie de Taylor de orden 5
# Expansiones de series de Taylor y otras funciones
sp.cos(x).series(x,0,12)       # Funcion coseno
sp.sin(x).series(x,0,9)       # Funcion seno
sp.exp(x).series()              #Funcion exponencial
(1 / ( x + 1 )).series()        #Función armónica

#Expanciones de funciones multivariadas.
expr = sp.cos(x) / (sp.sin(x * y) +1)
expr.series(x,n=4)
expr = expr.series(y,n=3)
```

```python
# CALCULO -> LÍMITES.
# Definición de funcion
f = sp.Function('f')
x, h = sp.symbols('x,h')
#Ejemplo
sp.limit(sp.sin(x) / x,x,0)         #lim_(x \to 0) \frac(sin(x)){x}
#Ejemplo
#Limites y derivadas
diff_limit=(f(x+h)-f(x))/h
# Asignamos a f una funcion dada.
sp.limit(diff_limit.subs(f,sp.cos), h, 0)       # derivada via limites del coseno
sp.limit(diff_limit.subs(f,sp.sin), h, 0)       # derivada via limites del seno
```
