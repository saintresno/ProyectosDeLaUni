```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,200)
y = np.sin(x)
#fig, ax= plt.subplots()
#ax.plot(x,y)
#plt.show()

x = np.linspace(-5,2,1000)
y1 = x**3+5*x**2+10
y2 = 3*x**2+10*x
y3 = 6*x+10

# Realizar en simultaneo la gráfica de las funciones

fig, ax = plt.subplots()
ax.plot(x,y1, color='blue', label= 'y(x)')
ax.plot(x,y2, color='purple', label= 'y´(x)')
ax.plot(x,y3, color='red', label= 'y´´(x)')
ax.legend()
plt.show()
```

```python
fig = plt.figure(figsize = (8,2.5), facecolor='#f1f1f1')
left = 0.1
bottom = 0.1
widht = 0.8
height = 0.8

#Manipulacion de olas coordenadas de los ejes
#Típicamente como fracciones del largo y ancho del lienzo (canvas)

ax = fig.add_axes((left, bottom, widht, height), facecolor='#f1f1f1' ) 
x = np.linspace(-2,2,1000)
y1 = np.cos(40*x)
y2 = np.exp(-x**2)
ax.plot(x,y1*y2)
ax.plot(x,y2, color = 'purple')
ax.plot(x,-y2, color = 'purple')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
plt.show()

#Guardar la figura
fig.savefig("Figura_1.png")
```

```python
x = np.linspace(-5,2,100)
y1 = x**3+5*x**2+10
y2 = 6*x**2+6*x
fig, ax = plt.subplots()
ax.hist(y1,20)
ax.hist(y2,20)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
plt.show()
```
