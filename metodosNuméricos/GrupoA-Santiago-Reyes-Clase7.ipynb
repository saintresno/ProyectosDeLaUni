```python
## ejemplo: Tiempo de vuelo

v0 = 5                          # velocidad inicial
g = 9.81                        #Gravedad
t = np.linspace(0,1,1000)      #vector tiempo
y = v0*t - 0.5*g*t**2

#altura máxima
largest_high = y[0]

for i in range(1,len(y),1):
    if y[i] > largest_high:
        largest_high = y[i]
    #primera línea después del if
#primera líena después del for
print ("la altura máxima alcanzada fue: ", largest_high)
```

```python
# figura altura tiempo.
plt.plot(t,y)
plt.xlabel('Time (S)')
plt.ylabel('height (m)')
plt.show
```

```python
## este es el script que simula una caminata aleatoria en 2D

import random
## parámetros.
N = 100            # número de pasos durante la caminata
d = 1               #longitud del paso.
#coordenadas.
X = np.zeros(N+1)           #coordenada x
Y = np.zeros(N+1)           #coordenada y

# posicion inicial

X[0] = 0; Y[0] = 0;

for i in range (0,N,1):
    r = random.random()         #Crea un número aleatorio entre 0 y 1
    #movimiento hacia arriba
    if (0<=r<=0.25):
        X[i+1] = X[i]
        Y[i+1] = Y[i]+d
    #movimiento hacia la derecha 
    elif (0.25<r<=0.5):
        X[i+1] = X[i]+d
        Y[i+1] = Y[i]
    #movimiento hacia la izquierda         
    elif (0.5< r<= 0.75):
        X[i+1] = X[i]-d
        Y[i+1] = Y[i]
    #movimiento hacia la abajo  
    else:
        X[i+1] = X[i]
        Y[i+1] = Y[i]-d


## figura
plt.plot(X,Y,'r--',X[0], Y[0],'bo', X[-1], Y[-1], 'b*')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
