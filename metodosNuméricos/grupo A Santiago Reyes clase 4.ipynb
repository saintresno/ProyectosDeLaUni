# Arreglos multidimencionales

```python
# arreglos multidimencionles
import numpy as np
```

```python
f = lambda m,n: n+10*m
A= np.fromfunction(f,(6,6), dtype= float)
A
```

```python
print (A[1,:], " la columna 0 \n") 
print (A[0,:], " la fila 0 \n")
print (A[:3,:3], " subloque matriz 3x3 superior-izquierdo \n") 
print (A[3:,:3], " subloque matriz 3x3 superior-derecho \n") 
print (A[:3,:6], " subloque matriz 3x6 \n") 
print (A[:3,3:], " primeras 3 columnas, últimas 3 filas \n") 
print (A[3:,:3], " últimas 3 columnas, primeras 3 filas \n") 
print (A[:,:], " Toda la matriz \n")
```

```python
print (A[::2, ::2], " elementos espaciados \n")
print (A[::2, ::3], " elementos espaciados \n")
print (A[1::2, 1::3], " elementos espaciados iniciando desde 1 espaciados dos unidades en fila y 3 en columnas \n")
print (A[1,2], " elemento con indice 1-2")
```

```python
B = A [1:5, 1:5]
print (B, " submatriz de posiciones 1 hasta 5 en indices \n")
B[:,:]=0
print (B, " \n")

print (A)
```

```python
C = B[1:3, 1:3].copy()
C[:,:] = 1,0
print (C , " \n \n", B)
```
