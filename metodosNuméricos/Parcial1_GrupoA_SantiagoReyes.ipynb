```python
import numpy as np

def raiz(P):
  x1 = P
  x2 = 0
  error = 0.1

  while(error > 0.00001):
    x2 = (x1 + (P/x1))/2 
    error = abs((x2-x1)/x1)
    x1 = x2
  return x2

print("La raiz cuadrada de 110 es: ", raiz(110))
print("La raiz cuadrada de 93.443 es: ", raiz(93.443))
print("La raiz cuadrada de 23.25 es: ", raiz(23.25))
```
