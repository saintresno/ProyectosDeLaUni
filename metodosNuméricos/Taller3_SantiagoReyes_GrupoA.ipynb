```python
import sympy as sp
sp.init_printing()
from sympy import pi, I, oo

phi=(1+sp.sqrt(5))/2
formula=(phi**4-phi)/(phi**7+1)
print("F = "+str(formula))
print("simplified F = "+str(sp.simplify(formula)))
```

```python
def InductionFormula(x,y):
    return (1+x)/y
a, b = sp.symbols('a b')
Sequence=[a,b]

print('u_0 = a')
print('u_1 = b')
for i in range(2,15):
    Sequence.append(sp.simplify(InductionFormula(Sequence[-1],Sequence[-2])))
    print('u_'+str(i)+' = '+str(Sequence[-1]))
```

```python
# Creamos una lista para almacenar las aproximaciones sucesivas
SuccessiveApproximations = [1]

# Definimos el número de iteraciones
p = 12

# Calculamos las sucesivas aproximaciones
for n in range(1, p):
    OldValue = SuccessiveApproximations[-1]
    NewValue = sp.expand(sp.sqrt(2 - 2 * sp.sqrt(1 - (OldValue ** 2) * sp.Rational(1, 4))))
    SuccessiveApproximations.append(NewValue)
    
    # Imprimimos la aproximación actual como una representación LaTeX y como un número decimal
    print(sp.latex(sp.simplify(3 * (2 ** n) * NewValue)))
    print(sp.N(NewValue*3*2**n))
```
