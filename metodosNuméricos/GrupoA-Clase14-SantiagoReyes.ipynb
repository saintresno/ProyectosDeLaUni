```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# series de datos
s = pd.Series([43728, 43728, 86547, 95482])
list(s.index)               #Lista de indices
s.values                    #Arreglo con los respectivos datos numéricos
```

```python
# Manipulación de índices

s.index=['Estocolmo', 'Londres', 'Roma', 'Paris']
s.name = 'poblacion'
s
```

```python
s = pd.Series([43728, 43728, 86547, 95482], name = 'poblacion', index=['Estocolmo', 'Londres', 'Roma', 'Paris'])
s['Estocolmo']
```

```python
# Resumen estadistico de la serie

s.mean(), s.median(), s.std()           # Media, Mediana y Desviacíon estandar
s.min(), s.max()                        #Minimo y máximo
s.quantile(q=0.25),s.quantile(q=0.5),s.quantile(q=0.75)     # primero, segundo y tercer cuartil.
```

```python
# Resumen estadistico simplificado
s.describe()
```

```python
#Exploración gráfica de los datos.

fig, axes = plt.subplots(1,4,figsize=(12,3))
s.plot(ax = axes[0], kind='line', title='Linea')
s.plot(ax = axes[1], kind='bar', title='Barras')
s.plot(ax = axes[2], kind='box', title='Caja')
s.plot(ax = axes[3], kind='pie', title='Torta')
```
