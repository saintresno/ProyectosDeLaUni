```python
x = np.linspace(-2,2,100)
#ejemplos de 4 funciones no lineales.
f1 = x**2 -x-1
f2 = x**3 -3*np.sin(x)
f3 = np.exp(x)
f4 = 1 - x**2 + np.sin(50/(1+x**2))
#grafica
fig,axes = plt.subplots(1,4,figsize=(12,3), sharey=True)
for n, f in enumerate([f1,f2,f3,f4]):
    axes[n].plot(x,f,lw = 1.5)
    axes[n].axhline(0,ls = ':', color = 'k')
    axes[n].set_ylim(-5,5)
    axes[n].set_xticks([-2,-1,0,1,2])
    axes[n].set_xlabel(r'$x$', fontsize = 18)
    axes[n].set_ylabel(r'$y$', fontsize = 18)
    titles = [r'$f(x)=X^2 - x-1$', r'$f(x)=x^3-3sin(x)', r'$f(x)= 0^x -2$', r'$1 - x^2 - sin(50/(1+x^2))$']
    for n, title in enumerate(titles):
        axes[n].set_title(title)
```

```python
from matplotlib.patches import ArrowStyle
#metodo de Biseccion
f = lambda x: np.exp(x) -2
tol = 0.1
a,b = -2,2
x = np.linspace(-1.9,2.1,1000)
# grafica 
fig, ax = plt.subplots(1,1, figsize=(12,4))
ax.plot(x, f(x), lw = 1.5)
ax.axhline(0,ls = ':', color = 'k')
ax.set_xticks([-2,-1,0,1,2])
ax.set_xlabel(r'$x$', fontsize = 18)
ax.set_ylabel(r'$f(x)$', fontsize = 18)

# método de Biseccion
fa,fb = f(a), f(b)
ax.plot(a,f(a),'ko')
ax.plot(a,f(b),'ko')
ax.text(a,f(a) + 0.5, r'$a$', ha='center', fontsize = 18)
ax.text(a,f(b) + 0.5, r'$b$', ha='center', fontsize = 18)
n = 1

while(b-a > tol):
    m = a + (b - a)/2
    fm = f(m)
    ax.plot(m,fm,'ko')
    ax.text(m,fm-0.5,r'$m_%d$' % n, ha = 'center')
    n+=1
    if (np.sign(fa) == np.sign(fm)):
        a, fa = m , fm
    else:
        b,fb = m,fm
ax.plot(m,fm,'r*', markersize = 10)
ax.annotate('Raiz aproximada en %.3f' % m, fontsize = 14, family = 'serif', xy = (a,fm),xycoords = 'data', xytext = (-150, +50), textcoords = 'offset points', arrowprops=dict(arrowstyle= '->', connectionstyle='arc3, rad=-.5'))
ax.set_title('Metodo de Biseccion')
```
