```python
df1 = pd.read_csv('/content/temperature_indoor_2014.tsv', delimiter = '\t', names = ['Tiempo', 'Afuera'])
df2 = pd.read_csv('/content/temperature_outdoor_2014.tsv', delimiter = '\t', names = ['Tiempo', 'Adentro'])
```
