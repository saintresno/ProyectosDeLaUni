```python
import sympy as smpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Recorrer PG desde 2 hasta 60mm Hg 
PG = np.linspace(2, 60, 100)

#Fijar el valor de AV con G = 4 & G = 5
AV_4 = 4 / np.sqrt(PG)
AV_5 = 5 / np.sqrt(PG)

#Mostrar o imprimir tabla
plt.plot(PG, AV_4, label='Q = 4 L/min')
plt.plot(PG, AV_5, label='Q = 5 L/min')
plt.xlabel('PG (mm Hg)')
plt.ylabel('AV (cm$^2$)')
plt.grid()
plt.legend()
plt.show()
```

```python
# Definir la ecuación para un perfil NACA simétrico de 4 dígitos
def NACA(x, c, t):
    y = (t * c / 0.2) * (0.2969 * np.sqrt(x / c) - 0.1260 * (x / c) \
         - 0.3516 * (x / c)**2 + 0.2843 * (x / c)**3 - 0.1015 * (x / c)**4)
    return y

# Definir las especificaciones de los perfiles NACA
c20 = 1.5 # Longitud de cuerda para NACA 0020
c33 = 2.0 # Longitud de cuerda para NACA 0033
t20 = 0.02 # Espesor relativo para NACA 0020
t33 = 0.033 # Espesor relativo para NACA 0033

# Crear una lista de puntos para trazar el perfil
x_points = np.linspace(0, 1, 200) * c33

# Calcular las coordenadas y para cada perfil
y1 = NACA(x_points, c20, t20)
y2 = NACA(x_points, c33, t33)

# Graficar los perfiles NACA
fig, ax = plt.subplots(figsize=(8,2))           #Tamaño del gráfico
ax.plot(x_points, y1, label='NACA 0020')        #Gráficar 1 y nombre de la línea
ax.plot(x_points, y2, label='NACA 0033')        #Gráficar 2 y nombre de la línea                    
ax.grid()                                       #Agregar cuadrícula a la gráfica
ax.legend()                                     #Agregar leyenda de los nombres de las gráficas
ax.set_xlabel('x/c')                            #Agregar nombre del eje x
ax.set_ylabel('y/c')                            #Agregar nombre del eje y
ax.set_title('Perfiles NACA simétricos')        #Agregar título de la gráfica
plt.show()                                      #Mostrar Gráfica
```
