```python
# altura promedio.

N = 5
h = np.zeros(N)
h[0] = 1.60; h[1] = 1.85; h[2] = 1.75; h[3] = 1.80; h[4] = 0.50;
suma = 0
for i in np.arange(0,5):
    suma = suma + h[i]
average = suma/N
print ('La altura promedio es : {:.1f} metros'.format(average))

```

```python
mean_high = np.mean(h)


print ('La altura promedio también es : ', mean_high, "metros")
```
