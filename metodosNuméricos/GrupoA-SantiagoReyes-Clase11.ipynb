```python
import sympy as sp
sp.init_printing()
from sympy import pi, I, oo
```

```python
x = sp.Symbol('x')

# atributos especiales 
# real, imaginary, positive, negative, integer, odd, even, primer, finito, infinite

x = sp.Symbol('x')
y = sp.Symbol('y', positive=True)

sp.sqrt(x**2)
sp.sqrt(y**2)
```

```python
n1=sp.Symbol('n')                   #representa un simbolo
n2=sp.Symbol('n', integer =True)    #representa un numero entero
n3=sp.Symbol('n', odd = True)       #representa un numero impar
#Calculamos el cos (n*pi)
sp.cos(pi*n1)
sp.cos(pi*n2)
sp.cos(pi*n3)
```

```python
# funciones 
x,y,z =  sp.symbols('x,y,z')

f = sp.Function('f')(x)
g = sp.Function('g')(x,y,z)
# para determinar los simbolos libres
g.free_symbols
```

```python
sp.sin(x)           #Definición general
sp.sin(1.5*pi)

n = sp.Symbol('n', integer=True)
sp.sin(n*pi)
```

```python
# funciones lambda

h = sp.Lambda(x,(x+1)**2)
h(5)
h(1+2*x)
```

```python
# Expreciones

expr= 1 + 2*x**2 + 3*x**3
expr_1 = 2 * (x**2-x) - x * (x+1)
sp.simplify(expr_1)
#Ejemplo
expr = 2 * sp.cos(x)*sp.sin(y)
sp.simplify(expr)
#ejemplo
expr = sp.exp(x) * sp.exp(y)
sp.simplify(expr)
```

```python
# expresiones
expr = (x+1) * (x+2)
sp.expand(expr)
```

```python
sp.sin(x+y).expand(trig = True)
expr = sp.sin(x+y)
sp.expand(expr,trig = True)
```

```python
a,b = sp.symbols('a,b', positive = True)
sp.log(a*b).expand( log=True)
```

```python
sp.exp(a*I+b).expand(complex = True)
```

```python
sp.expand((a*b) **x, power_base = True)
```

```python
sp.exp((a-b)*x).expand(power_exp=True)
```

```python
# factorizaciones
expr = x**2 -1
sp.factor(expr)
expr = x*sp.cos(y)+sp.sin(z)*x
sp.factor(expr)
```
