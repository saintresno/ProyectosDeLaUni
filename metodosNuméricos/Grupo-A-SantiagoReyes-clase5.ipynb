### operaciones entre matricez y vectores

```python
# Resuimen de funciones
# np.dot          Producnto punto entre vectores
# np.inner        Producto por escalar
# no.cross        Producto cruz entre vectores
# np.tensordot    Producto tensorial
# np.outer        Producto externo
# np.kron         Producto de kronecker

A = np.arange(1,7).reshape(2,3)
B = np.arange(1,7).reshape(3,2)
c = np.dot(A,B)
print ("tamaño 2x3 \n", A, "\n \n  tamaño 3x2 \n", B, " \n \n \n ", c)
```

```python
c = np.inner(2,A)
c 
```

```python
c = np.dot(B,A)
c
```
